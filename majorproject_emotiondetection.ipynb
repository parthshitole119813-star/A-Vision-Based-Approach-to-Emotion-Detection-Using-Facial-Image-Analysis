```python
# STEP 1: Install dependencies
!pip install deepface mediapipe opencv-python matplotlib

# STEP 2: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from deepface import DeepFace
from google.colab import files
from PIL import Image
import mediapipe as mp

# STEP 3: Upload an image
uploaded = files.upload()
image_name = list(uploaded.keys())[0]

# STEP 4: Read and display uploaded image
image = cv2.imread(image_name)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Uploaded Image')
plt.show()

# STEP 5: Detect all faces using MTCNN
detected_faces = DeepFace.extract_faces(img_path=image_name, detector_backend='mtcnn', enforce_detection=False)
print(f"✅ Detected {len(detected_faces)} face(s) in the image.\n")

# Initialize MediaPipe for landmarks (optional)
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=10, refine_landmarks=True)

# STEP 6: Analyze emotions for each detected face
for idx, face_data in enumerate(detected_faces, start=1):
    region = face_data["facial_area"]
    x, y, w, h = region['x'], region['y'], region['w'], region['h']

    # Crop face region
    face_crop = image_rgb[y:y+h, x:x+w]
    cv2.imwrite("temp_face.jpg", cv2.cvtColor(face_crop, cv2.COLOR_RGB2BGR))

    # Analyze this face
    result = DeepFace.analyze(img_path="temp_face.jpg", actions=['emotion'], enforce_detection=False)
    emotions = result[0]['emotion']
    sorted_emotions = sorted(emotions.items(), key=lambda x: x[1], reverse=True)
    top1, top2 = sorted_emotions[0][0], sorted_emotions[1][0]

    # Print emotion details
    print(f"🧠 Face #{idx} — Location: (x={x}, y={y}, w={w}, h={h})")
    print(f"   → Primary Emotion: {top1}")
    print(f"   → Secondary Emotion: {top2}")
    print("   Emotion Scores:")
    for emotion, score in emotions.items():
        print(f"     {emotion}: {score:.2f}%")
    print("--------------------------------------------------\n")

    # Draw bounding box and face number (no emotion text)
    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (0, 255, 0), 2)
    label = f"Face #{idx}"
    cv2.putText(image_rgb, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.8, (255, 0, 0), 2, cv2.LINE_AA)

    # Optional: facial landmarks (yellow points)
    results = face_mesh.process(face_crop)
    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            for lm in face_landmarks.landmark:
                lx, ly = int(lm.x * w) + x, int(lm.y * h) + y
                cv2.circle(image_rgb, (lx, ly), 1, (255, 255, 0), -1)

# STEP 7: Show final annotated image (numbers only)
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Detected Faces (Numbered Only)")
plt.show()

face_mesh.close()
```
